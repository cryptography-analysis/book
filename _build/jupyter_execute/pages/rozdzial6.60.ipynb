# Analiza Historyczna #
Żeby uzyskać szerszą perspektywę odnośnie szyfrów ADFGX/ADFGXVX, warto zapoznać się z kontekstem historycznym, skupiając się przede wszystkim na ogólnym obrazie niemieckich możliwości na polu kryptografii i kryptoanalizy w przeddzień I wojny światowej oraz w jej trakcie a także podejściu aliantów do tegoż zagadnienia.

##### **Braki w kryptoanalizie Niemców** #####
W „The Codebrakers”, David Kahn pisze: „Niemiecka kryptologia ślepo podążała w kierunku wojny wyposażona w ciężkie działa kryptografii i bez żadnej kryptoanalizy”. W 1914 roku niemiecka armia i marynarka były dobrze wyposażone w sprzęt bezprzewodowy jednak nie zdawali sobie sprawy z korzyści jak i niebezpieczeństw jakie niosła wówczas nowa technologia związana z komunikacją radiową. Ich dumne i sztywne usposobienie nie ułatwiało krytycznego myślenia i inicjatywy na niższych szczeblach. Podczas przygotowywań do wojny, dowództwo zabraniało swoim podwładnym na przechwytywanie komunikatów pochodzących od przeciwników. W marynarce wydarzył się nawet przypadek ukarania marynarza, który poinformował swoich przełożonych o przechwyceniu i odszyfrowaniu wiadomości wymienianych między dwoma punktami radiowymi. Generałowie Kaisera Wilhelma II (ówczesnego króla Prus) głęboko wierzyli w to, że zwyciężą w wojnie dzięki piechocie, artylerii i kawalerii tak jak swego czasu w zwycięskiej wojnie francusko-pruskiej w latach 1870-71. Bezpieczeństwo komunikacji czy łamanie szyfrów nie było rozważane jako coś istotnego.

Co więcej, niedługo po wybuchu wojny Niemcy ponieśli wielkie straty na polu rozpoznania elektromagnetycznego co miało w późniejszych latach niemałe konsekwencje. W sierpniu 1914, brytyjscy żołnierze z Royal Navy pocięli niemieckie kable telegraficzne znajdujące się pod wodą. Niemcy byli zmuszeni do przeorganizowania swojego systemu komunikacji i polegania bardziej na bezprzewodowych środkach co wiązało się z wystawieniem na niebezpieczeństwo przechwytywania komunikatów przez wroga. Dodatkowo, francuscy kryptologowie mieli tę przewagę nad swoimi niemieckimi kolegami po fachu, że posiadali w swoich szeregach grupę doświadczonych kryptografów, którzy obserwowali niemiecką armię jeszcze podczas jej manewrów przed wojną. Przewaga ta objawiła się już w początkowej fazie wojny, kiedy z sukcesem udała się im deszyfracja UBCHI - podstawowego kodu stosowanego przez Niemców co odegrało kluczową rolę na froncie wschodnim.

Oddziały niemieckie nie miały wpojonego używania sygnałów bezpieczeństwa więc często komunikowały się używając tekstu jawnego. Wywiad strategiczny Francji i Niemiec nie działał zbyt prężnie. Sztab Generalny Niemiec używał wywiadu wojskowego w zasadzie jedynie do pozyskiwania informacji dotyczących taktyki. Niemcy mieli ograniczone informacje na temat armii brytyjskiej z tego powodu, że jedynie marynarka zbierała informacje wywiadowcze dotyczące UK skupiając się przede wszystkim na Royal Navy. Kriegsmarine miała własną służbę wywiadu morskiego, Marinenachrichtendienst, która wykorzystywała namierzanie i ograniczone łamanie szyfrów do lokalizowania i identyfikowania okrętów wojennych wroga. Ministerstwo Spraw Zagranicznych posiadało własną siatkę wywiadowczą w swoich ambasadach i konsulatach, wraz w własnym biurem szyfrów – pracowano tam zwykle na źródłach jawnych i nie prowadzono współpracy z wojskowymi odpowiednikami. Aż do 1918 roku Ministerstwo wciąż uznawało swoje kody za nie do złamania. Podczas gdy Ministerstwo Wojny w Berlinie było odpowiedzialne za tworzenie kodów i szyfrów dla armii, nie istniała żadna instytucja zajmująca się jedynie kryptologią, która mogłaby być odpowiedzialna za zajęcie się kwestiami bezpieczeństwa.

W bitwie pod Tannenbergiem przeciwko dużo silniejszej armii rosyjskiej w 1914 roku, generał Paul von Hindenburg i jego kwatermistrz Ernst Ludendorff zwycięstwo zawdzięczali w dużej mierze parze amatorów w łamaniu szyfrów – porucznikowi Alexandrowi Beuermeisterowi oraz profesorowi filozofii Ludwigowi Deubnerowi. Deubner i Bauermeister zostali przydzieleni do Naczelnego Dowództwa Armii Wschodniej i współpracowali z biurem szyfrów austro-węgierskiego wywiadu wojskowego. Świeżo utworzony wywiad elektromagnetyczny w armii niemieckiej pozostawał ograniczony do frontu wschodniego. Dopiero później, gdy Hindenburg i Ludendorff stali na czele Naczelnego Dowództwa Armii, Oberste Heeresleitung (OHL), zaczęli integrować SIGINT na poziomie planowania strategicznego. Marynarka podążyła za nimi w lutym 1916 roku tworząc własną komórkę rozpoznania elektromagnetycznego i tworząc centrum przechwytywania i łamania kodów w Naumunster. Do tej pory jakiekolwiek sukcesy w łamaniu kodów, Niemcy zawdzięczali przypadkom inicjatywy pojedynczych jednostek lub błędów popełnianych przez wrogów.

Przełom nastąpił wiosną 1917 roku. Szef telegrafii polowej podlegający pod głównego kwatermistrza, został przemianowany na Nachrichtenchef co Childs przetłumaczył jako: „Główny Oficer Sygnałów”. Nachrichtenchef był odpowiedzialny za wszystkie formy komunikacji telefonicznej, telegraficznej, bezprzewodowej oraz używanie kodów i szyfrów.

Na szczeblu strategicznym kontrolę nad Abteilung IIIb zachował Szef Sztabu Generalnego Armii Polowej odpowiedzialny za planowanie operacyjne. Departament ten, oprócz klasycznego wywiadu wojskoweg, pełnił również funkcję policji politycznej w Niemczech oraz cenzora prasy. Ocena informacji wywiadowczych zebranych z SIGINT na temat armii wroga trafiała bezpośrednio do Nachrichtenabteilung, który w 1917 roku został przemianowany na Departament Wojsk Zagranicznych oraz posiadał dwie sekcje: wschodnią i zachodnią.

Według dawnego kryptologa Hermanna Stutzela Stacja Deszyfrująca w Naczelnym Dowództwie Armii miała trzy sekcje:
- Departament Wojsk Zagranicznych dokonujący oceny danych wywiadowczych przekazanych przez stacje przechwytujące. Skupili się na kodach okopów, ustalaniu współrzędnych oraz rozkazach bojowych wroga.
- Druga sekcja pracowała nad kodami polowymi dla armii niemieckiej
- Trzecia sekcja monitorowała ruch dyplomatyczny i pracowała nad łamaniem kodów i szyfrów dyplomatycznych. Na jej czele stanął właśnie Hermann Stutzel.

W 1918 roku Stutzel podjął inicjatywę monitorowania ruchu między MSZ a jego ambasadą w Madrycie i ujawnił podatność niemieckich kodów dyplomatycznych. Jednak pomimo skandalu jaki wybuchł po opublikowaniu telegramu Zimmermanna (który w 1917 roku wciągnął USA do wojny z państwami centralnymi) i pomimo pracy Stutzela, MSZ wciąż utrzymywało, że jego kody są nie do złamania. Twierdziło również, że wyciek wiadomości Zimmermanna był spowodowany zdradą.

Podobna pewność co do bezpieczeństwa swoich kodów towarzyszyła również kryptografom Niemieckiej Armii. Wśród nich był porucznik Fritz Nebel, który służył w Wielkim Sztabie Generalnym w Spa. Specjalizował się on w grece oraz językach latynoskich jednak w momencie dołączania do batalionu telegraficznego, nie miał jakiegokolwiek przeszkolenia w zakresie kryptografii. Niemniej na początku 1918 roku otrzymał polecenie opracowania nowego, wysoko bezpiecznego szyfru polowego, który był potrzebny Niemcom do nadciągającej wiosennej ofensywy. Nebel opracował szyfr ADFGX, który później rozszerzył poprzez dodanie większej ilości symboli (ADFGVX).

Wiele lat później w 1960 roku na zjeździe dawnych kryptologów Niebel pierwszy raz spotkał Georgesa Poinvina, swojego byłego francuskiego oponenta. Był zszokowany kiedy Poinvin wyjaśnił mu sposób w jaki złamał szyfr ADFGVX oraz przechwyconą w maju 1918 roku wiadomość zawierającą istotne informacje strategiczne o niemieckich planach, mianowicie „Radiogram zwycięstwa”. Umożliwiło to aliantom zatrzymanie ofensywsy niemieckiej i odzyskanie inicjatywy na froncie zachodnim. Nebel wyznał później, że nie był „ani matematykiem ani inżynierem” w czasie opracowywania kodów. Owe zeznania potwierdzają surowy werdykt Kahna na temat niemieckiej kryptologii w czasach I wojny światowej.


##### **James Rives Childs i Kryptoanaliza Aliantów** #####
Sukcesy francuzów w łamaniu niemieckich kodów na początku wojny nie były przypadkiem. Po porażce we francusko-pruskiej wojnie na przełomie lat 1870/71, ywiad francuski ciężko pracował nad łamaniem niemieckich kodów. Wiedza ta umożliwiła francuskiemu wywiadowi wojskowemu interpretację i deszyfrację zakodowanych niemieckich wiadomości od samego początku wojny. Francuskie dowództwo zdołało przejąć rozkaz bojowy wroga i dowiedzieć się o planie ataku na Paryż oraz zatrzymać ofensywę niemiecką na zachodzie („cud nad Marną”).

Po tych wydarzeniach francuscy kryptoanalitycy połączyli siły z brytyjskimi sojusznikami a od 1917 roku również z amerykanami. Wymieniali między sobą odszyfrowane wiadomości, wiedzę i metody. W USA do Sztabu Generalnego Amerykańskich Sił Ekspedycyjnych przydzielono Dywizję Wywiadu Radiowego pod dowództwem ppłk. Franka Moormana, która obejmowała sekcję przechwytywania radiowego oraz sekcję rozwiązywania kodów, na której czele stał William F. Friedman.

Jednym z oficerów podlegających Friedmanowi był James Rives Child. Child dołączył do Armii Amerykańskiej w 1915 roku po ukończeniu studiów z wyróżnieniem. Służył w Amerykańskich siłach ekspedycyjnych i został przydzielony do służby przechwytywania radiowego jako oficer łącznikowy z francuskimi i brytyjskimi służbami łamania szyfrów. Po wojnie opracował ogólne rozwiązanie szyfru ADFGVX oraz otrzymał medal wolności za pracą nad łamaniem niemieckich szyfrów. 

W 1934 roku Wydział Planów Wojennych i Szkolenia Stanów Zjednoczonych opublikował „Ogólne rozwiązanie systemu szyfrującego ADFGVX”, które zostało już odtajnione.


##### **Odczytywanie niemieckich komunikatów z Rumunii.** #####
Po klęsce z Niemcami, 7 maja 1918 roku Rumunia podpisała „pokój w Bukareszcie”. Na jego mocy , większość rumuńskiej armii musiała został zdemobilizowana a kraj miał być zajęty przez niemieckie wojska okupacyjne dowodzone przez feldmarszałka Augusta von Mackensena. Początkowo ich kwatera znajdowała się w Bukareszcie lecz później została przeniesiona do Pałacu Palescha (120km na północ od Bukaresztu). Mackensen wydawał się doceniać nowoczesne technologie jak telefon i telegraf, o czym mówił później w ten sposób: „Wojna rozwinęła i udoskonaliła te udogodnienia. Możemy porozumiewać się z centralą zupełnie tak, jakbyśmy siedzieli naprzeciwko siebie.”

Rozległa sieć kolejowa łączyła okupowaną Rumunię z Cesarstwem Niemieckim oraz portem w Konstancji nad Morzem Czarnym. Kolejnym punktem o znaczeniu logistycznym i strategicznym był dla Niemiec port w Mikołajowie (Ukraina). Każdego dnia 100 wagonów z ropą wyruszało do Niemiec, które niezwykle tego potrzebowały do swojej machiny wojennej.

Wynik wojny zależał jednak przede wszystkim od rozwoju sytuacji na froncie zachodnim. Ofensywy niemieckie podczas 1918 roku zakończyły się klęską, również dzięki sukcesom aliantów na polu kryptoanalitycznym. W październiku imperium Austro-Węgierskie zaczęło rozpadać się na mniejsze narody. Niektóre z nich, jak Bułgaria, wycofały się z sojuszu z państwami centralnymi i podpisały rozejm z Ententą. Podobnie postąpiły następnie Węgry a ostatecznie 3 listopada również Austria. Tego samego dnia wybuchł bunt w niemieckiej marynarce wojennej, który rozprzestrzenił się na całe Cesarstwo Niemieckie i kraje okupowane. Brytyjczycy przechwycili wówczas 13-częściową wiadomość, która następnie została odszyfrowana przez Childsa. Wiadomość ta zwierała plany generała Mackensena na wycofanie się z Rumunii.

Ta długa wiadomość została wysłana z kwatery głównej Mackensena to Naczelnego Dowództwa Armii. Jej odszyfrowanie umożliwiło aliantom wgląd w plany Mackensena oraz jego spojrzenie na sytuację. Według Kahna, odszyfrowana depesza została natychmiast przekazana do Francuskiej Najwyższej Rady Wojennej.

9 listopada, cesarz został zmuszony do abdykacji a Niemcy stały się republiką. Dwa dni później rząd niemiecki podpisał traktat pokojowy w Compiegne i rozpoczął wycofywanie swoich wojsk z frontu zachodniego. Wojna zakończyła się, jednak setki tysięcy niemieckich żołnierzy wciąż przebywało na froncie wschodnim, odcięci od swojej ojczyzny. Z powodu braku odpowiedniej ilości zasobów (węgla, wagonów etc.) nie było możliwości przeprowadzenia zorganizowanej ewakuacji. Sytuacja była krytyczna ponieważ dotrzymanie terminów, które narzuciła ententa było niemożliwe.

25 listopada 1918 r. o godzinie 16.30 Mackensen wysłał do Naczelnego Dowództwa w Berlinie kolejną wiadomość, która również została przechwycona i odszyfrowana przez Childsa. Mackensen informował Naczelne Dowództwo Armii, Ministerstwo Wojny oraz nowy rząd w Berlinie, że francuski dowódca w regionie generał Berthelot, zawiadomił go, że generał Foch odmawia zastosowania się do warunków rozejmu na froncie południowo-wschodnim i żąda aby wojska Mackensena zostały natychmiast rozbrojone i internowane. Mackensen pisał również, że czeka na decyzję Naczelnego Dowództwa, zanim potwierdzi otrzymanie wiadomości od generała Berthelota. Zasugerował również aby rozbrojenie i internowanie dokonało się na terenie Węgier a nie krajów Ententy ponieważ to zapewniłoby żołnierzom odpowiednie zaopatrzenie w żywność. Obawiał się, że ta sytuacja może przerodzić się w „masowy exodus z nieopisaną nędzą i otwartą rewoltą”.

Czekając na ostateczną decyzję, Mackensen zaczynał coraz bardziej martwić się o komunikację. Nie mógł już korzystać z Nikołajewa (NKJ) jako stacji przekaźnikowej a z Berlinem komunikował się poprzez stację Breslau (BSL). Twierdził, że radio zapewnia „wyjątkowo bezpieczną łączność z ojczyzną”.

28 listopada alianci odszyfrowali wiadomość przesłaną z Berlina poprzez Breslau do Mackensena z instrukcjami aby „unikać jakiejkolwiek walki i zastosować się do wymagań Francji. Czekać na kolejne rozkazy.”. Ostateczne instrukcje przybyły 29 listopada i ta wiadomość również została odszyfrowana przez Childsa. Los armii Mackensena był przesądzony:

Ostatnia wiadomość z kolekcji Childsa dotycząca Rumunii datuje się na 30 listopada 1918 roku. Niezidentyfikowany oficer imieniem Papp, używający znaku wywoławczego CV, prosi 11 Armię o „przesłanie na czas przepisów dotyczących internowania”. W grudniu feldmarszałek wkroczył na Węgry ze swoimi pozostałymi wojskami. Został internowany w pałacu w komfortowych warunkach i pilnowany przez Francuzów a większość jego żołnierzy została odesłana do Niemiec.


##### **Rewolucja październikowa w Niemczech** #####

28 października 1918 roku w Kilonii wybuchło powstanie cywilno-wojskowe przeciwko wojnie oraz rządowi cesarskiemu. Załogi Cesarskiej Marynarki Wojennej odmówiły wykonania rozkazu dotyczącego ostatniego ataku na brytyjską marynarkę. Początkowo władze wojskowe bezskutecznie usiłowały stłumić bunt. 3 listopada bunt przerodził się w rebelię, która rozprzestrzeniła się po całym Imperium Niemieckim i państwach okupowanych. Niemiecki rząd i Naczelne Dowództwo Armii wciąż próbowały zatrzymać rozprzestrzenianie się buntu. W związku z tym byli zmuszeni do zabezpieczenia infrastruktury komunikacyjnej aby uniknąć przejęcia stacji radiowych przez rebeliantów.
6 listopada o godzinie 13:31 Naczelny Oficer Łączności wysyła rozkaz określony jako „bardzo pilny” z głównej radiostacji wojskowej w Berlinie do radiostacji wojskowych w Konstantynopolu i Mikołajewie a za ich pośrednictwem do wszystkich stacji radiowych w rejonie Morza Czarnego, w tym do Rumunii, Gruzji i Turcji. Wszystkie radiostacje powinny zostać przejęte a odpowiedzialny za nie oficer ma ich strzec „jak oka w głowie” oraz uniemożliwiać nagrywanie i rozpowszechnianie wiadomości radiowych z Kilonii. Wcześniej otrzymane wiadomości miały zostać skonfiskowane i zniszczone. Stacja w Mikołajewie przesłała wiadomość dalej a pierwsza część transmisji została przechwycona i 7 listopada odszyfrowana przez Childsa.
8 listopada biuro łączności poinformowało Sztab Generalny w Spa (SY), że niemiecka radiostacja wojskowa w Warszawie „poddała się Radzie Żołnierskiej”. Dzień później oficer łączności poinformował SY, że od godziny 22:00 radiostacja będzie pod kontrolą Rady Żołnierzy ale na razie wciąż „trzyma w zaciśniętej pięści” zarówno radiostację jak i personel.
9 listopada oficjalnie proklamowano Republikę Niemiecką w Berlinie. Tego samego dnia nowa wiadomość została przechwycona i odszyfrowana przez Childsa.  Zawierała opis rozprzestrzeniania się Rewolucji znad Bałtyku do zachodnich granic Imperium Niemieckiego. Dywizja sygnałowa informowała Dowództwo w Spa, że Rada Żołnierzy w Kolonii próbuje przejąć lokalną stację radiową. Kolonia wraz ze swoją infrastrukturą była istotna pod względem strategicznym zarówno dla Marynarki jak i Armii. Starając się w dalszym ciągu zachować kontrolę nad komunikacją, Naczelne Dowództwo wydało rozkaz – który również został odszyfrowany przez aliantów – aby zaprzestać używania sieci kablowej (telefonicznej i telegraficznej) a posługiwać się jedynie transmisją radiową i wojskowymi szyframi. Ponadto wszystkie komunikaty miały od tamtej pory być potwierdzane dwukrotnie. 
Bez wątpienia Rewolucja Listopadowa miała ogromny wpływ na chęci i pośpiech Niemców do podpisania rozejmu. Aliantom udawało się śledzić przygotowania do podróży delegacji niemieckiej oraz zidentyfikować jednostkę kierującą operacją. 
11 listopada o 19:06 wojskowa radiostacja w Berlinie wysłała szczegóły zawieszenie broni do Mikołajewa z poleceniem przekazania ich „za wszelką cenę” do Dywizji Śródziemnomorskiej i do Dowództwa Specjalnego Operacyjnego Marynarki Wojennej. Wiadomość ta została prawdopodobnie wysłana dość późno gdyż mowa w niej o tym, że zawieszenie broni miało obowiązywać od południa tego samego dnia.
Jednym z warunków zawieszenia broni była całkowita demilitaryzacja regionu między Renem a zachodnią granicą Niemiec w wyniku czego, dowództwo przeniosło się do Kassel. Jednakże rewolucja dotarła również do tego miasta. 10 listopada niżsi stopniem żołnierze zaczęli rozbrajać swoich oficerów. Powstała Rada Robotników i Żołnierzy a nad miejskim łukiem triumfalnym zaczęła powiewać czerwona flaga. Podczas próby jej ściągnięcia jeden z oficerów zostaje postrzelony. Opis tych wydarzeń również znalazł się w przechwyconych wiadomościach ADFGVX. W wiadomości z 12 listopada pewien porucznik Bauman (prawdopodobnie oficer łączności kierujący łąznością radiową), otrzymał polecenie „by był gotowy do jak najszybszego przemieszczenia się do Kassel wojskowym transportem”.
13 listopada Rada Żołnierska w Warszawie zwróciła się z zapytaniem do radiostacji w Mikołąjewie czy plotki odnośnie wkroczenia feldmarszałka Mackensena wraz z wojskami do Warszawy są prawdą. Mackensen był szeroko znany jako zagorzały zwolennik monarchii a Rada Żołnierska obawiała się, że przejdzie on przez Polskę ze swoimi 170 000 żołnierzami podczas drogi z Rumunii do Niemiec.


##### **Kod w kodzie** #####

Trzy wiadomości podzielone na cztery kryptogramy, zawierają enigmatyczne odniesienia do niezidentyfikowanych osób czy podmiotów takich jak „Dyktator” lub „Herkules”. Pierwsza wiadomość została przesłana 4 listopada o 14:05 z Berlina do Bukaresztu. Urząd wojenny informuje personel ekonomiczny Mackensena, że „przybył transport złota” oraz, że „Herkules” ponownie odrzuca odpowiedzialność za nowe połączenia alarmowe od „Andreas”.
Tego samego dnia o 19:56 kolejna wiadomość została wysłana ze stacji radiowej Marynarki Wojennej w Eilwes do Osmani, głównej radiostacji w Konstantynopolu. Wiadomość zawiera trzy części, które znajdują się na oddzielnych stronach w książce Childsa – pierwsza na stronia 120 a druga i trzecia na 123. Nadawcą jest dział zaopatrzenia Rzeszy Niemieckiej. Informuje, że „Dyktator” przekazał instrukcje aby „nie dzwonić pod numer alarmowy do Andreas” oraz, że ponowny kontakt został nawiązany z „Herkulesem”. Ta enigmatyczna osoba, odrzuca wszystkie „nowe połączenia alarmowe”. 
Trzecia wiadomość pochodzi z dnia 15 listopada, została przesłana o godzinie 16:50 i podpisana przez departament zaopatzenia oraz admirała Staffa. Wysłano ją z głównej stacji radiowej w Berlinie do radiostacji o nowym sygnale wywoławczym (XYZ), prawdopodobnie była to radiostacja wojskowa w Konstantynopolu. Zgodnie z tą wiadomością, „Herkules” poinformował Berlina, że 12 listopada kilka parowców z „Operette” dotarło do Mikołajewa w blisko 10 tonami wątróbki i kakao na pokładzie. „Ładunek pozostaje pod wojskową strażą” z powodu problemów z transportem. 
Możemy jedynie spekulować odnośnie prawdziwych tożsamości – „Dyktator” może oznaczać szefa Naczelnego Dowództwa von Hindenburga. „Andreas” to prawdopodobnie kryptonim Gruzji a „Herkules” może oznaczać Naczelne Dowództwo lub Dowództwo Marynarki Wojennej zaś „Operette” port w rejonie Konstantynopola. Jednakże wciąż nie jest jasne dlaczego w ogóle zastosowano kryptonimy w tych wiadomościach skoro w większości innych nie były używane.
